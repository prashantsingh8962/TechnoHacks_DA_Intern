# Task 4 : Pivote Table

## We have titanic dataset for this task. It consist of two tables Train and Test.

Use pivot tables to summarize data in a dataset

## Train table : 

```python
import pandas as pd
```

```python
# reading train table
df1= pd.read_csv("C:\\Users\\Prashant Singh\\OneDrive\\Documents\\Internships\\TechnoHacks\\train.csv")
df1
```

```python
# Average fare paid by passengers based on class and gender
pivot_table = df1.pivot_table(values='Fare', index='Pclass', columns='Sex', aggfunc='mean')
print(round(pivot_table,2))
```

```python
# Average Survived paid by passengers based on class and gender
pivot_table = df1.pivot_table(values='Fare', index='Survived', columns='Sex', aggfunc='mean')
print(round(pivot_table,2))
```
