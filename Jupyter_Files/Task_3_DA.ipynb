# Task 3 : Visualization using Histogram

## We have titanic dataset for this task. It consist of two tables Train and Test.

Create a histogram or bar chart to visualize the distribution of data in a dataset

## Train table :

```python
import pandas as pd
```

```python
# reading train table
df1= pd.read_csv("C:\\Users\\Prashant Singh\\OneDrive\\Documents\\Internships\\TechnoHacks\\train.csv")
df1
```

```python
df1.info()
```

```python
import matplotlib.pyplot as plt

# Create a histogram for the Age column
plt.figure(figsize=(10, 6))
plt.hist(df1['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
# Create a bar chart for Pclass distribution
plt.figure(figsize=(10, 6))
df1['Pclass'].value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Passenger Class Distribution')
plt.xlabel('Pclass')
plt.ylabel('Number of Passengers')
plt.show()
```
